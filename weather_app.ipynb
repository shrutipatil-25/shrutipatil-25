```python
import requests
import gradio as gr

def weather(city):
    api_key = "fd8b6effa849be1dd606cc8f6b353858"
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    try:
        res = requests.get(url)
        res.raise_for_status()  # Raise an error for bad responses (404, 401, etc.)
        data = res.json()
        print(data)  # Debugging print

        # Check if the request was successful
        if data["cod"] == 200:
            temp = data["main"]["temp"]
            description = data["weather"][0]["description"]  # Corrected key
            return f"The temperature in {city} is {temp}°C\nDescription: {description.capitalize()}"
        else:
            return "City not found. Please enter a valid city name."

    except requests.exceptions.RequestException as e:
        return f"Error fetching data: {e}"

# Gradio Interface
interface = gr.Interface(
    fn=weather,
    inputs=gr.Textbox(lines=2, placeholder="Enter City name"),
    outputs="text"
)

interface.launch()
```
